Görüntüde tarif edilen içerik, URL parametrelerinin bir HTTP GET isteği içinde nasıl kullanılacağını açıklıyor. GET istekleri, genellikle bir web sunucusundan veri almak için kullanılır ve URL parametreleri, sunucunun isteğe nasıl yanıt vereceğini özelleştirmek için ek bilgi sağlamak amacıyla kullanılır.

Parametreler, URL'nin "query string" bölümünde yer alır. 
 - Bu bölüm, URL'nin sonunda, bir soru işareti (?) ile başlar 
 - ve ardından anahtar-değer çiftleri gelir. 
 - Her çift, eşittir işareti (=) ile ayrılır 
 - ve çiftler arası ampersand (&) işareti ile ayrılır.
 
Bu URL'de:

- name parametresinin değeri Joseph'tur.
- ID parametresinin değeri 123'tür.
Bu iki parametre, sunucuya kimin veya hangi verinin istendiğini bildirir ve sunucu bu parametrelere göre uygun yanıtı döndürür. 

Python requests kütüphanesinde, bu tip parametreleri bir sözlük olarak 
 - params argümanı ile 
 - requests.get() 
 fonksiyonuna verebilirsiniz

 Burada requests kütüphanesinin get fonksiyonunun imzası ve açıklaması yer alıyor. get fonksiyonu, bir HTTP GET isteği göndermek için kullanılır ve bir web sunucusundan veri almak için standart bir yol sunar. İşte detayları:

url: İsteğin gönderileceği tam URL adresi. Bu, istenen kaynağın internet üzerindeki konumudur.

params: Bu parametre, URL'nin sorgu dizgisine (query string) eklenecek olan ekstra bilgileri içerir. Genellikle bir sözlük, bir liste çiftleri veya bayt verisi olarak sağlanır. Örneğin, params={'name': 'Joseph', 'ID': '123'} sözlüğü kullanıldığında, bu anahtar-değer çiftleri URL'ye ?name=Joseph&ID=123 şeklinde eklenir.

**kwargs: Bu, requests.request fonksiyonunun alabileceği diğer tüm isteğe özgü seçenekleri ve ayarları içerir. Bu parametreler arasında zaman aşımı (timeout), oturum (session), çerezler (cookies), HTTP başlıkları (headers) ve daha fazlası yer alabilir.

Fonksiyonun dönüş değeri bir requests.Response nesnesidir. Bu nesne, sunucudan alınan yanıtın tüm bilgilerini içerir ve durum kodu (status code), yanıt başlıkları (headers), yanıt içeriği (content) ve daha fazlasını erişilebilir kılar.

## Post Requests

Like a GET request, a POST is usef to send data to server, but the POST request sends the data in a request body. In order to send the Post Request in Python, in the URL we change the route to POST.

Gönderdiğiniz ekran görüntüsü, HTTP POST isteğinin nasıl yapıldığını ve Python `requests` kütüphanesi ile bu isteğin nasıl işlendiğini açıklıyor. İşte adım adım açıklaması:

1. Bir POST isteği, sunucuya veri göndermek için kullanılır. Veriler, HTTP isteğinin gövdesinde (request body) gönderilir.

2. Gösterilen örnekte `http://httpbin.org/post` adresine bir POST isteği yapılıyor. `httpbin.org`, HTTP isteklerini test etmek için tasarlanmış bir hizmettir ve gönderilen istekleri çeşitli şekillerde yansıtarak, HTTP isteklerini test etmek için kullanılır.

3. `payload` adlı bir sözlük (dictionary) oluşturuluyor. Bu sözlük, POST isteğiyle birlikte gönderilecek verileri içeriyor. Örneğin, `payload = {"name": "Joseph", "ID": "123"}` gibi.

4. `requests.post(url_post, data=payload)` fonksiyon çağrısı ile bu veriler `data` parametresi kullanılarak POST isteği gövdesinde sunucuya gönderiliyor.

5. POST isteğinin yanıtı, `r_post` değişkeninde saklanıyor. Bu yanıtın URL'si ve gövdesi (`request body`), POST isteğinin detaylarını göstermek için kullanılıyor.

6. `r_post.url` ve `r.url` ifadeleri, POST ve GET isteklerinin URL'lerini karşılaştırmak için yazdırılıyor. GET isteğinin URL'sinde sorgu parametreleri URL'nin kendisine eklenirken, POST isteğinde veriler URL'de değil, istek gövdesinde gönderilir.

7. `r_post.request.body` ve `r.request.body` ifadeleri, POST ve GET isteklerinin gövdelerini (request bodies) karşılaştırmak için yazdırılıyor. POST isteğinin bir gövdesi olurken (verileri içerir), GET isteğinin tipik olarak bir gövdesi olmaz.

8. `r_post.json()['form']` ifadesi, sunucudan gelen yanıttaki form verilerini JSON formatında almak için kullanılıyor. `httpbin` yanıt olarak gönderilen verileri `form` anahtarında JSON olarak geri yollar.

Bu örnekte, ekran görüntüsünde sağlanan bilgiler, temel HTTP isteklerini ve yanıtlarını anlamanıza yardımcı olur ve POST ile GET istekleri arasındaki farkları gösterir. `requests` kütüphanesi, bu tür HTTP işlemlerini kolaylıkla gerçekleştirmek için oldukça yararlıdır.

## Passing Parameters In URLs

You often want to send some sort of data in the URL’s query string. If you were constructing the URL by hand, this data would be given as key/value pairs in the URL after a question mark, e.g. httpbin.org/get?key=val. Requests allows you to provide these arguments as a dictionary of strings, using the params keyword argument. As an example, if you wanted to pass key1=value1 and key2=value2 to httpbin.org/get, you would use the following code:

```python
payload = {'key1': 'value1', 'key2': 'value2'}
r = requests.get('https://httpbin.org/get', params=payload)
```

```python
# see that the URL
print(r.url)
```

```python
# Note that any dictionary key whose value is None will not be added to the URL’s query string.

# You can also pass a list of items as a value:

payload = {'key1': 'value1', 'key2': ['value2', 'value3']}

r = requests.get('https://httpbin.org/get', params=payload)
print(r.url)
```
